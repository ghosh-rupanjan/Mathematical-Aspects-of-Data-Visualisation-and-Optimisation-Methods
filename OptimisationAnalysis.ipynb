# Task 1

DEFINING FUNCTIONS

```python
def rosenbrock(X):
  f = (1-X[0])**2+100*(X[1]-X[0]**2)**2
  return f

def rastrigin(X):
  f = 10*len(X)
  for i in range (0,len(X)):
    f += X[i]**2 - 10*np.cos(2*np.pi*X[i])
  return f

def ackley(X):
  f = 20+ np.e
  k1, k2 = 0,0
  for i in range (0,len(X)):
    k1 += X[i]**2
    k2 += np.cos(2*np.pi*X[i])
  f += -20*np.exp(-0.2*np.sqrt(k1/len(X)))-np.exp(k2/len(X))
  return f
```

```python
#DEFINING THE VECTOR ON WHICH THE FUNCTION ROSENBROCK ACTS
X = [0.5,0.5]
#Nelder-Mead, simulated annealing and CMAE optimisation paths
result_NM = minimize(rosenbrock, X , method = "Nelder-Mead")
result_sa = dual_annealing(rosenbrock, bounds = [(-3,2),(-3,2)])
cma_1 = cma.CMAEvolutionStrategy(X,0.3)
cma_1.optimize(rosenbrock)
result_cma = cma_1.result
print(result_NM,result_sa,"\n", result_cma)
```

```python
#DEFINING THE VECTOR ON WHICH THE FUNCTION RASTRIGIN ACTS
X = [0.1,0.1]
#Nelder-Mead, simulated annealing and CMAE optimisation paths
result_NM = minimize(rastrigin, X , method = "Nelder-Mead")
result_sa = dual_annealing(rastrigin, bounds = [(-3,2),(-3,2)])
cma_1 = cma.CMAEvolutionStrategy(X,0.3)
cma_1.optimize(rastrigin)
result_cma = cma_1.result
print(result_NM,result_sa,"\n", result_cma)
```

```python
#DEFINING THE VECTOR ON WHICH THE FUNCTION ACKLEY ACTS
X = [0.5,0.3]
result_NM = minimize(ackley, X , method = "Nelder-Mead")
result_sa = dual_annealing(ackley, bounds = [(-3,2),(-3,2)])
cma_1 = cma.CMAEvolutionStrategy(X,0.3)
cma_1.optimize(ackley)
result_cma = cma_1.result
print(result_NM,result_sa,"\n", result_cma)
```

# Task 2

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version = 1)

# take out the data and labels, only 500 elements taken for implementation
X = mnist["data"][0:500]
Y = mnist["target"][0:500]
X = X.to_numpy()
Y = Y.to_numpy()
# split the data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(X_train[5])
```

```python
#defining kernel array for optimisation
k = ["linear","poly","rbf","sigmoid"]
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
```

```python
#defining the regulation parameter(cv), gamma(yv), kernel index(kv) and accurarcy(accv) as global lists
cv,yv,kv,accv = [],[],[],[]
k1 = []
def svm_opt(arr,X_train,Y_train,Y_test,X_test):
  [k_idx, C, y ] = arr
  #print(k_idx,C,y)
  clf = make_pipeline(StandardScaler(), svm.SVC())
  clf = svm.SVC(C = C, gamma = y, kernel = k[int(round(k_idx,0))])
  clf.fit(X_train,Y_train)

  cv.append(C)
  yv.append(y)
  kv.append(int(round(k_idx,0)))
  accv.append(round(-accuracy_score(Y_test,clf.predict(X_test)),9))
  #returned -accuracy to get minimum negetive accuracy
  return -accuracy_score(Y_test,clf.predict(X_test))

#defining bounds for the variables
bnd = [(0, 3), (0, 10), (0, 5)]
```

# Nelder-Mead optimisation(Task 2 and 3)

```python
result_NM = minimize(svm_opt,[2,3,2],args = (X_train,Y_train,Y_test,X_test),method = "Nelder-Mead", bounds = bnd)
print(result_NM)
```

```python
cv,yv,kv,accv = np.array(cv),np.array(yv),np.array(kv),np.array(accv)
print(accv)
fig = plt.figure()
ax = fig.add_subplot(111,projection = "3d")
scatter = ax.scatter(cv,yv,kv,c = accv,cmap = 'coolwarm', alpha = 0.6)
ax.plot(cv,yv,kv,color ="k", alpha = 0.2)
```

Robustness to initial condition

```python
result_NM = minimize(svm_opt,[1,5,1],args = (X_train,Y_train,Y_test,X_test),method = "Nelder-Mead", bounds = bnd)
print(result_NM)
cv,yv,kv,accv = np.array(cv),np.array(yv),np.array(kv),np.array(accv)
print(accv)
fig = plt.figure()
ax = fig.add_subplot(111,projection = "3d")
scatter = ax.scatter(cv,yv,kv,c = accv,cmap = 'coolwarm', alpha = 0.6)
ax.plot(cv,yv,kv,color ="k", alpha = 0.2)
```

# CMAE optimisation(Task 3)

```python
cma_1 = cma.CMAEvolutionStrategy([0,8,2],0.3)
cma_1.optimize(svm_opt,args = (X_train,Y_train,Y_test,X_test))
result_cma = cma_1.result
print(result_cma)
```

```python
cv,yv,kv,accv = np.array(cv),np.array(yv),np.array(kv),np.array(accv)
print(accv)
fig = plt.figure()
ax = fig.add_subplot(111,projection = "3d")
scatter = ax.scatter(cv,yv,kv,c = accv,cmap = 'coolwarm', alpha = 0.6)
ax.plot(cv,yv,kv,color ="k", alpha = 0.7)
ax.set_xlabel("C")
ax.set_ylabel("gamma")
ax.set_zlabel("kernel")
```
